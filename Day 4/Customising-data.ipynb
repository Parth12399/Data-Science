# Making our Data Attractive

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
heart_disease = pd.read_csv('heart-disease.csv')
heart_disease
```

```python
heart_disease.head()
```

```python
over_50 = heart_disease[heart_disease['age'] > 50]
over_50.head()
```

```python
len(over_50)
```

```python
len(heart_disease)
```

```python
plt.style.available
```

## Initial Plot

```python
fig, (ax0,ax1) = plt.subplots(nrows = 2, ncols = 1,
                             figsize = (10,10), sharex = True)
#Add data to ax0
scatter = ax0.scatter(x=over_50['age'],y=over_50['chol'],
                      c=over_50['target'])

#Customising ax0
ax0.set(title = 'Heart-Diseases And Cholestrol level',
        ylabel = 'Cholestrol')

#Add a legend
ax0.legend(*scatter.legend_elements(),title = 'Target')

#Add a horizontal line
ax0.axhline(heart_disease['chol'].mean(),linestyle = '--')

#Add data to ax1
scatter = ax1.scatter(x=heart_disease['age'],y=heart_disease['thalach'],
                     c=heart_disease['target'])

#Customising ax1
ax1.set(title = 'Heart-Diseases And Max Heart rate ',
       xlabel = 'Age',ylabel = 'Max Heart rate')

#Add a legend
ax1.legend(*scatter.legend_elements(),title = 'Target')

#Add a horizontal line
ax1.axhline(heart_disease['thalach'].mean(),linestyle = '--')

#Add title to figure
fig.suptitle('Heart Disease Analysis', fontsize = 16, fontweight = 'bold');
```

## Final Plot

```python
#Styling background
plt.style.use('seaborn-whitegrid')

fig, (ax0,ax1) = plt.subplots(nrows = 2, ncols = 1,
                             figsize = (10,10), sharex = True)
#Add data to ax0
scatter = ax0.scatter(x=over_50['age'],y=over_50['chol'],
                      c=over_50['target'],
                      cmap = 'winter')

#Customising ax0
ax0.set(title = 'Heart-Diseases And Cholestrol level',
        ylabel = 'Cholestrol')
ax0.set_xlim((50,80))

#Add a legend
ax0.legend(*scatter.legend_elements(),title = 'Target')

#Add a horizontal line
ax0.axhline(heart_disease['chol'].mean(),linestyle = '--')

#Add data to ax1
scatter = ax1.scatter(x=heart_disease['age'],y=heart_disease['thalach'],
                     c=heart_disease['target'],
                     cmap = 'winter')

#Customising ax1
ax1.set(title = 'Heart-Diseases And Max Heart rate ',
       xlabel = 'Age',ylabel = 'Max Heart rate')
ax1.set_xlim((50,80))
ax1.set_ylim((60,200))
#Add a legend
ax1.legend(*scatter.legend_elements(),title = 'Target')

#Add a horizontal line
ax1.axhline(heart_disease['thalach'].mean(),linestyle = '--')

#Add title to figure
fig.suptitle('Heart Disease Analysis', fontsize = 16, fontweight = 'bold');
```
