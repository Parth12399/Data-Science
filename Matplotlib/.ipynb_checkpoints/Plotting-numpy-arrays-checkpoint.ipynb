# Making figures using numpy

We will plot:
* Line plot
* Scatter plot
* Bar graph
* Histogram
* Subplots

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,100,10)
```

```python
x
```

```python
#line plot
fig,ax = plt.subplots()
ax.plot(x,x**2);
```

```python
#Scatter plot
fig,ax = plt.subplots()
ax.scatter(x,np.sin(x))
```

```python
#Bar Graph
#Preparing Data and then plotting
Chocolate_prices = {'Silk':120,'Silk Oreo':150,'Bubbly':50}
fig,ax = plt.subplots()
ax.bar(Chocolate_prices.keys(),Chocolate_prices.values());
```

```python
#Horizontal bar graph
fig,ax = plt.subplots()
ax.barh(list(Chocolate_prices.keys()),list(Chocolate_prices.values())) 
```

```python
#Histogram
fig,ax = plt.subplots()
ax.hist(np.random.randn(100));
```

```python
#Creating Subplots
fig ,((ax1,ax2),(ax3,ax4)) = plt.subplots(nrows = 2 , ncols = 2,
                                         figsize = (10,10))
ax1.plot(np.random.randn(100))
ax2.scatter(x,np.exp(x))
ax3.bar(Chocolate_prices.keys(),Chocolate_prices.values())
ax4.hist(np.random.randint(20,size=(5,3)));
```

```python
#Creating Subplots method 2
fig, ax = plt.subplots(nrows = 2, ncols = 2,
                      figsize = (10,10))
ax[0,0].plot(np.random.randn(100))
ax[0,1].scatter(x,np.exp(x))
ax[1,0].bar(Chocolate_prices.keys(),Chocolate_prices.values())
ax[1,1].hist(np.random.randint(20,size=(5,3)));
```
